```python
import matplotlib.pyplot as plt
import numpy as np
```

# End-To-End Memory Networks
$ l_{kj} = (1−j/J)−(k/d)(1−2j/J) $

```python
# 단어 수
J = 15

# 벡터 차원
d = 20

l = np.zeros([J, d], dtype=np.float32)

for j in range(J):
    j += 1 # 1-based indexing
    for k in range(d):
        k += 1 # 1-based indexing
        l[j-1,k-1] = (1-j/J) - (k/d)*(1-2*j/J)
```

```python
fig = plt.figure(figsize=(10, 5))
fig.patch.set_facecolor('white')
plt.plot(l)
plt.xlabel('word position')
plt.ylabel('value')
fig
```

# Attention is all you need

$PE_{(pos, 2i)} = sin(pos/10000^{2i/d_{model}})$

$PE_{(pos, 2i+1)} = cos(pos/10000^{2i/d_{model}})$

```python
PE = np.zeros([J, d], dtype=np.float32)

for pos in range(J):
    pos += 1
    for i in range(d):
        i += 1
        if i % 2 == 0:
            PE[pos-1, i-1] = np.sin(pos / np.power(10000, (2*i/d)))
        else:
            PE[pos-1, i-1] = np.cos(pos / np.power(10000, (2*i/d)))
```

```python
fig = plt.figure(figsize=(10, 5))
plt.plot(PE)
plt.xlabel('word position')
plt.ylabel('value')
fig
```
